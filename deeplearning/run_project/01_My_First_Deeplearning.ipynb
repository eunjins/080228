```python
# -*- coding: utf-8 -*-
# 코드 내부에 한글을 사용가능 하게 해주는 부분

# 딥러닝을 구동하는 데 필요한 케라스 함수를 불러옴
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# 필요한 라이브러리를 불러옴
import numpy as np
import tensorflow as tf
```

```python
# 실행할 때마다 같은 결과를 출력하기 위해 설정
np.random.seed(3)
tf.random.set_seed(3)

# 준비된 수술 환자 데이터를 불러들임
Data_set = np.loadtxt("../dataset/ThoraricSurgery.csv", delimiter=",")
```

```python
# 환자의 기록과 수술 결과를 X와 Y로 구분하여 저장
X = Data_set[:,0:17]
Y = Data_set[:,17]
```

## 딥러닝 구조 결정(모델 설정 및 실행)

model 함수를 선언하면 딥러닝 모델 설정 및 구동 시작

딥러닝: 퍼셉트론 위에 숨겨진 퍼셉트론 층을 차곡차곡 추가

구현 함수: 케라스 Sequential()

Sequential() 함수를 model로 선언하고, model.add()를 하면 새로운 층(은닉층, 출력층) 생성
- 출력층: 맨 마지막 층
- 은닉층: 나머지
- 입력층은 따로 만들지 않음

첫번째 은닉층에 input_dim을 적으면 첫번째 Dense가 은닉층+입력층 역할을 함

```python
# 딥러닝 구조를 짜고 층을 설정
model = Sequential()

# 두 개의 층을 가진 모델을 만듦 (model.add)
# 데이터에서 17개의 값을 받아 은닉층의 30개 노드로 보낸다. 활성화 함수는 relu를 사용한다. 
model.add(Dense(30, input_dim=17, activation='relu'))
# 출력층(마지막층) 생성. 출력층의 노드는 1개, 노드에서 입력받은 값은 sigmoid 활성화 함수를 거쳐 최종 출력 값으로 나온다.
model.add(Dense(1, activation='sigmoid'))
```

## 딥러닝 실행


```python
# 이항 교차 엔트로피 함수(binary_crossentropy) 사용
# metrics: 모델이 컴파일될 떄 모델 수행 결과를 나타내게 설정. 과적합 문제(overfitting) 방지
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
# model.fit(): 컴파일 환경에서 주어진 데이터를 불러 실행
# 샘플이 처음부터 끝까지 100번 재사용될 때까지 실행을 반복한다. 샘플은 한 번에 10개씩 끊어서 집어넣는다.
model.fit(X, Y, epochs=100, batch_size=10)
```
